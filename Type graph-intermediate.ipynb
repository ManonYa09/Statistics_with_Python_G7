## 3. Pie chart

is a circle that is divided into sections or wedges according to the
percentage of frequencies in each category of the distribution.


````python
df = pd.read_csv('Dataset/Bank Customer Churn Prediction.csv')
````

````python
df
````

````python
df.value_counts()
````

````python
df[['country']].value_counts()
````

````python
labels=df['country'].value_counts().index
````

````python
fig, ax = plt.subplots(figsize = (6, 6), dpi = 100)
ax.pie(x = df['country'].value_counts(), 
       labels=df['country'].value_counts().index,
      autopct='%1.2f%%',
      explode= [0, 0.1, 0],
      startangle=270)
ax.legend(df['country'].value_counts().index)
plt.savefig('Pie_chart.png', dpi = 1000)
plt.show()
````

````python
fig, ax = plt.subplots( figsize = (6, 6), dpi = 100)
ax.pie(x = df['country'].value_counts(),
       labels = df['country'].value_counts().index,
      autopct='%1.1f%%',
      explode= [0.1, 0, 00],
      startangle=90)
ax.set_title('Distribution of Country')
ax.legend(df['country'].value_counts().index, loc='upper left')
fig.savefig('pie.png', dpi = 1000)
plt.show()
````

## 5. Histrogram

**Histrogram** is a graph that shows the frequency of numerical data using rectangles. The height of a rectangle (the vertical axis) represents the distribution frequency of a variable (the amount, or how often that variable appears).

**Histogram** is a graph that displays the data by using contiguous vertical bars
(unless the frequency of a class is 0) of various heights to represent the
frequencies of the classes.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

## 6. Measures of position: quartiles
### Box Plot
<a id='box_plot'></a>
**Box plots** provide a quick visual summary of the variability of values in a dataset. They show the median, upper and lower quartiles, minimum and maximum values, and any outliers in the dataset. Outliers can reveal mistakes or unusual occurrences in data.

````python
fig, ax = plt.subplots(figsize = (4, 10))
sns.boxplot(df['estimated_salary'])
ax.set_yticks(np.arange(10000, 2.1*(10**5), 10000))
ax.grid(False)
````

## Violin Plot
<a id='Violin_plots'></a>

### How to Interpret Violin Plots?

Similar to `histograms`, `violin plots` show the shape of a data set. However, unlike histograms, violin plots show the shape of a data set by using a Probability Density Function (PDF), or a density plot, which is effectively a smoothed-over histogram. 


<img src="Photos/violin-plot-describe.png" alt="Control Structure" width="50%" style="display: block; margin: 0 auto;">

````python
fig, ax = plt.subplots(figsize = (4, 6))
sns.violinplot(df['estimated_salary'])
ax.set_yticks(np.arange(10000, 2.1*(10**5), 10000))
ax.grid(False)
````
