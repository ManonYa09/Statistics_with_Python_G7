# Visualizing Frequency Distributions

The data set is about basketball players in WNBA (Women's National Basketball Association), and contains general information about players, along with their metrics for the season 2016-2017.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Bar Plots

```python
wnba = pd.read_csv("DataSet/wnba.csv")
```

```python
wnba['Pos'].value_counts().plot.bar()
```

```python
wnba['Pos'].value_counts().plot.barh(title = 'Number of players in WNBA by position')
```

### Pie Charts

```python
wnba['Pos'].value_counts().plot.pie()
```

```python
wnba['Pos'].value_counts().plot.pie(figsize = (6,6))
plt.ylabel("")
```

```python
wnba['Pos'].value_counts().plot.pie(figsize = (6,6), autopct = '%.1f%%')
```

### Hostograms

```python
wnba['PTS'].plot.hist()
```

```python
wnba['PTS'].value_counts(bins = 10).sort_index()
```

```python
from numpy import arange
wnba['PTS'].plot.hist(grid = True, xticks = arange(2,585,58.2), rot = 30)
```

```python
wnba['PTS'].describe()
```

```python
wnba['Games Played'].plot.hist()
```

```python
wnba['Games Played'].describe()
```

```python
wnba['PTS'].plot.hist(range = (0,600), bins = 3)
```

```python
wnba['PTS'].plot.hist(range = (0,600), bins = 6)
```

```python
wnba['Games Played'].plot.hist(range=(0,32), bins=8, title='The distribution of players by games played', xticks=range(0,33,4))
plt.xlabel('Games played')
```
